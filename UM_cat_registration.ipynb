```python
#@title load model
import sys
sys.path.append('yolov7_bbox')
from face_alignment import *
import yaml
import torch
from yolov7_bbox.models2.yolo import Model
from yolov7_bbox.bbox_detection import detect as bbox_detect

import sys
sys.path.append('yolov7_bbox_landmarks')

from yolov7_bbox_landmarks.models.experimental import attempt_load
from yolov7_bbox_landmarks.kpts_detection import detect as kpt_detect

import iresnet
import torch
from torch import nn
import shutil

import cv2
import numpy as np
from PIL import Image
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt

import clip
import gdown
import shutil
import os

@torch.no_grad()
def inference(net, img):
    if img is None:
        img = np.random.randint(0, 255, size=(112, 112, 3), dtype=np.uint8)
    else:
        img = cv2.resize(img, (112, 112))

    img = np.transpose(img, (2, 0, 1))
    img = torch.from_numpy(img.copy()).unsqueeze(0).float()
    img.div_(255).sub_(0.5).div_(0.5)
    net.eval()
    with torch.no_grad():
        feat = net(img.to('cuda'))
    return feat

def save_yaml(data, file_path):
    with open(file_path, 'w') as file:
        yaml.dump(data, file)

def load_yaml(file_path):
    with open(file_path, 'r') as file:
        data = yaml.safe_load(file)
    return data

def pet_face_detector(img_path):
    img = cv2.imread(img_path)
    imgs = []
    _, bbox_list, src_pts_list, conf = kpt_detect(yolo_model1, img_path, device)
    _, bbox_list2, conf2 = bbox_detect(yolo_model2, img_path, device)
    if src_pts_list and bbox_list:
        for i, (bbox, s) in enumerate(zip(bbox_list, src_pts_list)):
            src_pts = adjust_src_pts(np.array(s).reshape(3, 2))
            face_img = preprocess(img, bbox, src_pts, image_size="112, 112", margin=0)
            imgs.append(face_img)
    elif bbox_list2:
        for i, bbox in enumerate(bbox_list2):
            crop_img = img[int(bbox[1]):int(bbox[3]), int(bbox[0]):int(bbox[2])]
            imgs.append(crop_img)
    return imgs


file_path = 'yolov7_bbox/yolo_model2_yaml.yaml'
yolo_model2_yaml = load_yaml(file_path)

weights = 'yolov7_bbox/weights/yolov7x-pet-face.pth'
state_dict = torch.load(weights)
state_dict = {key[6:]: value for key, value in state_dict.items()}

device = 'cuda' if torch.cuda.is_available() else 'cpu'
yolo_model2 = Model(yolo_model2_yaml).to(device)
yolo_model2.load_state_dict(state_dict)
yolo_model2.autoshape()

device = 'cuda' if torch.cuda.is_available() else 'cpu'
weights = 'yolov7_bbox_landmarks/weights/yolov7-pet-face.pt'
yolo_model1 = attempt_load(weights, map_location=device)

clip.available_models()
device = "cuda" if torch.cuda.is_available() else 'cpu'
clip_model, clip_preprocess = clip.load("ViT-B/16")

model_path = 'best_data25_2024_12_31.pth'
device = 'cuda'
backbone = iresnet.iresnet100(num_classes=512)
backbone.load_state_dict(torch.load(model_path, map_location=torch.device(device)))
backbone.to(device)

from PIL import Image

def clip_clf(img, text, model=clip_model):
    image = clip_preprocess(img).unsqueeze(0).to(device)

    with torch.no_grad():
        I_f = model.encode_image(image)
        T_f = model.encode_text(text)

    logits_I, logits_T = model(image, text)
    probs = logits_I.softmax(dim=-1).detach().cpu().numpy()

    return probs

text_pool = [
    "it is an image of a dog",
    "it is an image of a cat",
    "a cat face",
    "a dog face",
    "it is a background image",
    "it is an image of human",
    "it is an advertisement",
]

text = clip.tokenize(text_pool).to(device)
```

```python
#@title 把暫存路徑的相片格式由heic轉換為jpg
from PIL import Image
from pillow_heif import register_heif_opener

register_heif_opener()
import os
import shutil
from tqdm.notebook import tqdm

file_path = f'../drive/MyDrive/{upload_file_name}'

for name in tqdm(os.listdir(file_path)):
  if 'heic' in name.lower():
    image = Image.open(f'{file_path}/{name}')
    save_name = ''.join(name.split('.')[:-1]) + '.jpg'
    image.save(f'{file_path}/{save_name}')
    os.remove(f'{file_path}/{name}')
    print(f'convert: {name} --> {save_name}')
```

```python
#@title face detection
raw_imgs_path = [f'{upload_file_path}/{name}' for name in os.listdir(upload_file_path) if 'jpg' in name.lower() or 'png' in name.lower()]

print(f'total images: {len(raw_imgs_path)}')

print('face detecion:')
dst1 = upload_file_path

total_faces = 0
os.makedirs(dst1, exist_ok=True)
q_embs = None
for raw_img_path in raw_imgs_path:
  if 'face_align' in raw_img_path:
    continue
  imgs = pet_face_detector(raw_img_path)
  img_name = raw_img_path.split('/')[-1]

  for i, img in enumerate(imgs):
    emb = inference(backbone, img)
    q_embs = torch.vstack((q_embs, emb)) if q_embs is not None else emb
    cv2.imwrite(f'{dst1}/face_align_{i}_{img_name}', img)
    total_faces += 1
print(f'😺:Total faces: {total_faces}')

print('\n🐈:Loading database')
db_file_path = f'../drive/MyDrive/{file_name}'
src = db_file_path

cos = nn.CosineSimilarity(eps=1e-6).to(device)

img_paths = []
labels = []
for folder in os.listdir(src):
    if ('xlsx' in folder) or ('cache' in folder):
      continue

    for img_path in os.listdir(f"{src}/{folder}"):
        if 'face_align' in img_path:
            img_paths.append(f"{src}/{folder}/{img_path}")
            labels.append(folder)
labels = np.array(labels)

for i, img_path in enumerate(tqdm(img_paths)):
    img = cv2.imread(img_path)
    img = cv2.resize(img, (112, 112))
    emb = inference(backbone, img)
    embs = emb if i == 0 else torch.vstack((embs, emb))
print('🐈:Done!')
```

# 對比database照片的相似度

```python
#df_stat["品種"].unique()
```

```python
#@title 從database 找出最相似的貓, 如果登記的是第一隻貓, 請省略這一步
import cv2

def add_bound(img, line=1, color=(0, 0, 0)):
    bg = np.ones((112 + line*2, 112 + line*2, 3)) * color
    bg = bg.astype(np.uint8)
    bg[line:-line, line:-line] = img
    return bg

def plot_fa_grid(path, color=(0, 0, 0)):
  lst = []
  for img_name in os.listdir(path):
    if 'face_align' not in img_name:
        continue
    img = cv2.imread(f'{path}/{img_name}')[..., ::-1]
    img = cv2.resize(img, (112, 112))
    lst.append(add_bound(img, color=color))
  grid = np.hstack(lst)
  return grid


db_file_path = f'../drive/MyDrive/{file_name}'

threshold = 0.3 #@param {type:"slider", min:0.3, max:1.0, step:0.01}

use_breed_and_gender_for_filtering = False # @param {"type":"boolean","placeholder":"額外使用品種和性別過濾"}
breed = "" # @param {type:"string", placeholder:"輸入貓的品種"}
gender = 'M' # @param ["M", "F"]

filter_labels = []
if df_stat is not None and use_breed_and_gender_for_filtering:
   filter_labels = df_stat[(df_stat['品種'] == breed) & (df_stat['性別'] == gender)]['名字'].to_list()

%matplotlib inline
print('🐱: 新登記的貓')
plt.imshow(plot_fa_grid(upload_file_path, color=(255, 0, 0)))
plt.axis('off')
plt.show()

if embs.shape[0] > 0:
  # 對比
  sim_id_list = []
  sim_scores_list = []
  for q in q_embs:
    cos_sims = cos(q.reshape(1, -1), embs)
    mask = (cos_sims > threshold)
    most_sim_id = labels[mask.cpu().numpy()]
    sim_scores = cos_sims[mask]
    #print(cos_sims)
    sim_id_list += most_sim_id.tolist()
    sim_scores_list += sim_scores.tolist()

  sim_id_list = np.array(sim_id_list)
  sim_scores_list = np.array(sim_scores_list)
  inds = np.argsort(sim_scores_list)[::-1]

  sim_id_list_uni = []
  sim_scores_list_uni = []
  if sim_id_list.size > 0:

    for id_, sim_ in zip(sim_id_list[inds], sim_scores_list[inds]):
      if id_ in sim_id_list_uni:
        continue

      if use_breed_and_gender_for_filtering:
        if id_ not in filter_labels:
          continue

      sim_id_list_uni.append(id_)
      sim_scores_list_uni.append(sim_)
      print(f'🐈: 貓的姓名: {id_}, 相似度: {sim_}')
      plt.imshow(plot_fa_grid(f'{db_file_path}/{id_}'))
      plt.axis('off')
      plt.show()
  else:
    print(f'🐈: 沒有找到相似的貓')
else:
  print(f'🐈: 沒有找到相似的貓')
```
